## üîÅ List Comprehensions

```python
# 1. What does the list contain?
X = [i for i in range(100)]
print(X)
```

```python
# 2. S‚ÇÅ = {x in X | x % 5 == 2}
S1 = [x for x in X if x % 5 == 2]
print(S1)
```

```python
# 3. S‚ÇÇ = {x in S‚ÇÅ | x is even}
S2 = [x for x in S1 if x % 2 == 0]
print(S2)
```

```python
# 4. Generate all tuples (x, y) where x in S‚ÇÅ and y in S‚ÇÇ
pairs = [(x, y) for x in S1 for y in S2]
print(pairs[:10])  # Print first 10 pairs
```
